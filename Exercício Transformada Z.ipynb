```python
# Definição de funções comuns e bibliotecas
import numpy as np
import matplotlib.pyplot as plt

def u(n):
    if n >= 0:
        return 1
    else:
        return 0

n = np.arange(-10, 11) # determina o valor de n os números de -10 à 11
u_valores = np.vectorize(u)(n) # Aplica a cada loop um valor de u para cada n 
```

```python
# Plotagem do gráfico de u(n)

plt.stem(n,u_valores)
plt.xlabel('n')
plt.ylabel('u(n)')
plt.title('Gráfico da função u(n)')
plt.grid()
plt.show()
```

```python
# Defina o círculo unitário
theta = np.linspace(0, 2*np.pi, 100)
x = np.cos(theta)
y = np.sin(theta)

# Plotar círculo unitário
plt.figure()
plt.plot(x, y, '-', linewidth=2)
plt.axis('equal')
plt.axis([-1.2, 1.2, -1.2, 1.2])
plt.title('Círculo Unitário')
plt.grid(True)

# Plotar eixo x e eixo y
plt.xlim(-1.5, 1.5)
plt.ylim(-1.5, 1.5)
plt.axhline(0, color='black', lw=1.25)
plt.axvline(0, color='black', lw=1.25)

# Plotar polos e zeros
num = [1]
den = [1, -1]
zeros = np.roots(num)
poles = np.roots(den)
plt.plot(poles.real, poles.imag, 'x', markersize=12, label='Polos')
plt.plot(zeros.real, zeros.imag, 'o', markersize=10, label='Zeros')
plt.legend()
plt.show()
```
